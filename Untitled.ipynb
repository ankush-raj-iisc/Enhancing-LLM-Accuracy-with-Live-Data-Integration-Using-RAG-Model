```python
import requests

def google_search(query, api_key, search_engine_id):
    url = f"https://www.googleapis.com/customsearch/v1?key={api_key}&cx={search_engine_id}&q={query}"
    response = requests.get(url)
    return response.json()  # Returns the search result as JSON

# Replace the placeholders with your actual API Key and Search Engine ID
api_key = "API-KEY"
search_engine_id = "Search_engine-ID"
query = "Identify the industry in which Canoo (Canoo, a publicly traded company listed on NASDAQ) operates, along with its size, growth rate, trends, and key players."

# Perform the search
results = google_search(query, api_key, search_engine_id)
print(results)
```
